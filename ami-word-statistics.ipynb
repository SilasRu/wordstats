```python
import os
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('ami_words.csv')
len(df)
```

```python
def plot_means(n):
    top = df.groupby('words').count().sort_values(by='time_delta', ascending=False).iloc[:n+1]
    for i in range(0,n, 100):
        sorted_means = df[df['words'].isin(top.index)].groupby('words').mean().sort_values(by='time_delta').iloc[i:i+100]
        sorted_means.plot.bar(figsize=(30,5), title=f'means of top {i}:{i+100} spoken words', ylim=(0,1))

def check_wordlength(word):
    if word in df['words'].values:
        word_matches = df[df['words']== word]
        print(word_matches.describe())
    else:
        print('Word does not exist in library')
```

## Plot means
---

```python
plot_means(500)
```

## Check individual words
---

```python
check_wordlength('this')
```
